## <font color = blue size  = 5> **Task 3: Linear Regression (3 Marks)**

**Task Description:**
- Use the following helper functions in your notebook to compute the regression line
- Then answer the questions

```python
def standard_units(arr):
    return (arr - np.mean(arr)) / np.std(arr)
```

```python
def correlation(t, x, y):
    return np.mean(standard_units(t.column(x)) * standard_units(t.column(y)))
```

```python
def slope(t, x, y):
    r = correlation(t, x, y)
    return r * np.std(t.column(y)) / np.std(t.column(x))
```

```python
def intercept(t, x, y):
    return np.mean(t.column(y)) - slope(t, x, y) * np.mean(t.column(x))
```

```python
# Q6. Using the helper functions above, compute and print the slope and intercept of the regression line that predicts petal_width from petal_length. Write the full regression equation.

# Using the helper functions to compute slope and intercept
def standard_units(arr):
    return (arr - np.mean(arr)) / np.std(arr)

def correlation(t, x, y):
    return np.mean(standard_units(t[x]) * standard_units(t[y]))

def slope(t, x, y):
    r = correlation(t, x, y)
    return r * np.std(t[y]) / np.std(t[x])

def intercept(t, x, y):
    return np.mean(t[y]) - slope(t, x, y) * np.mean(t[x])

# Compute the slope (m) and intercept (b)
m = slope(data, 'PetalLengthCm', 'PetalWidthCm')
b = intercept(data, 'PetalLengthCm', 'PetalWidthCm')

# Print the regression equation
print("Regression Equation:")
print("y =", m, "x +", b)
```

```python
# Q7. Use your regression equation to predict the petal width of a flower with petal_length = 4.5 cm. Show your calculation.

# Use the regression equation to predict petal width for petal length = 4.5 cm
petal_length = 4.5
predicted_petal_width = m * petal_length + b

# Show the result
print("Predicted Petal Width for Petal Length of", petal_length, "cm is:")
print(predicted_petal_width, "cm")
```

```python
# Q8. Compute the correlation coefficient r between petal_length and petal_width. Is the relationship strong or weak? Positive or negative? Explain in 2-3 sentences.

# Calculate correlation coefficient between PetalLengthCm and PetalWidthCm
r = correlation(data, "PetalLengthCm", "PetalWidthCm")

# Print the correlation coefficient
print("Correlation coefficient (r):")
print(r)
print("")
print("Observation: ")
print("The correlation coefficient r between petal length and petal width is around 0.96, which is very close to 1. This indicates a strong positive linear relationship between the two variables.")
```

## <font color = blue size  = 5> **Task 4: Residual Analysis (4 Marks)**


- Use the regression line you calculated in the following questions.

```python
# Q9. Compute the predicted petal_width for every flower using your regression equation. Add both the predicted values and the residuals (actual - predicted) as new columns to the table. Display the first 8 rows.

# Use the regression equation: y = mx + b
predicted_petal_width = m * data["PetalLengthCm"] + b

# Calculate the residuals (actual - predicted)
residuals = data["PetalWidthCm"] - predicted_petal_width

# Add the predicted values and residuals as new columns
data["PredictedPetalWidth"] = predicted_petal_width
data["Residuals"] = residuals

# Display the first 8 rows
print(data[['PetalLengthCm', 'PetalWidthCm', 'PredictedPetalWidth', 'Residuals']].head(8))
```

```python
# Q10. Create a scatter plot of the residuals (y-axis) vs petal_length (x-axis). Add a horizontal red dashed line at y = 0. In 2-3 sentences, interpret what this plot tells you about how well the linear model fits the data.

# Scatter plot of residuals vs petal length
plt.scatter(data["PetalLengthCm"], data["Residuals"])

# Add a title
plt.title("Residuals vs Petal Length")

# Add labels
plt.xlabel("Petal Length (cm)")
plt.ylabel("Residuals")

# Add horizontal red dashed line at y = 0 (zero residuals)
plt.axhline(y=0, color='red', linestyle='--', linewidth=2)

# Show the plot
plt.show()

# Print the observation
print("")
print("Observation: ")
print("")
print("""For petal lengths below 4 cm:
The residuals are fairly small but spread out between -0.3 and +0.35, indicating the regression model has moderate prediction errors for smaller flowers.""")
print("")
print("""For petal lengths above 4 cm:
The residuals vary more between -0.6 and +0.65, showing larger errors in prediction and a less accurate fit for the larger flowers. """)

```
